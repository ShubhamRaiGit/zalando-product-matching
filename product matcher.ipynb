```python

stop_words = stopwords.words("german")
stop_words = set(stop_words)
def remove_stop_words(text):
    word_tokens = word_tokenize(text)
    words_tokens_lower = [word.lower() for word in word_tokens]
    words_filtered = [
        word for word in words_tokens_lower if word not in stop_words
    ]

    text_clean = " ".join(words_filtered)
    return text_clean
```

```python
def remove_punct(x):
    if isinstance(x, str):
        punc = '''!()-[]{};:'"\,<>./?|@#$%^&*_~'''
        for ele in x:
            if ele in punc:
                x = x.replace(ele, " ")
        return re.sub(' +', ' ', x)
    else:
        return ''
```

##### 3.1) Create Matches Pairs for Aboutyou and Zalando and assign label 1

```python
def get_offer(products, match, shop):
    return products[products['offer_id'] == match[shop]].iloc[0]


def create_dataset():
    i = 0
    product = []
    while i <= len(matches_training_df) - 1:
        temp_prod1 = get_offer(offers_training_df, matches_training_df.iloc[i],
                               'zalando')
        temp_prod1 = temp_prod1[[
            'offer_id', 'shop', 'brand', 'color', 'title', 'price',
            'description', 'image_urls'
        ]]
        index_name = [
            'ltable_offer_id', 'ltable_shop', 'ltable_brand', 'ltable_color',
            'ltable_title', 'ltable_price', 'ltable_description',
            'ltable_image_urls'
        ]
        temp_prod1.index = index_name

        temp_prod2 = get_offer(offers_training_df, matches_training_df.iloc[i],
                               'aboutyou')
        temp_prod2 = temp_prod2[[
            'offer_id', 'shop', 'brand', 'color', 'title', 'price',
            'description', 'image_urls'
        ]]
        index_name = [
            'rtable_offer_id', 'rtable_shop', 'rtable_brand', 'rtable_color',
            'rtable_title', 'rtable_price', 'rtable_description',
            'rtable_image_urls'
        ]
        temp_prod2.index = index_name
        final_df = pd.concat([temp_prod1, temp_prod2], axis=0)
        product.append(final_df)
        i += 1
    product = pd.DataFrame(product)
    return product


final_dataset_matches = create_dataset()

final_dataset_matches['label'] = 1
```

```python
zalando_matches = final_dataset_matches['ltable_offer_id']
about_you_matches = final_dataset_matches['rtable_offer_id']
print(len(zalando_matches))
print(len(about_you_matches))
union_matches = pd.concat([zalando_matches, about_you_matches])

union = pd.Series(np.union1d(offers_training_df['offer_id'], union_matches))
intersect = pd.Series(
    np.intersect1d(offers_training_df['offer_id'], union_matches))
no_matches = union[~union.isin(intersect)]

no_matches_dataset = offers_training_df[offers_training_df.offer_id.isin(
    no_matches)]
matches_dataset = offers_training_df[offers_training_df.offer_id.isin(
    intersect)]

group1 = no_matches_dataset.loc[no_matches_dataset.shop == 'aboutyou']
group2 = no_matches_dataset.loc[no_matches_dataset.shop == 'zalando']
```

##### 3.2) Create overlap blocker on Title and Brand to form pairs from zalando and aboutyou which are _not in_ Matches files and assign label 0.</br>
__Note:We are using overlap blocker instead of randomly assigning pairs for no matches to imporve quality of train dataset. Since we will be removing obvious mismatches from final submission through blockers there is no need to train dataset on obvious mismatches.__

```python
zal = group2[[
    'offer_id', 'shop', 'brand', 'color', 'title', 'price', 'description',
    'image_urls'
]]
abt = group1[[
    'offer_id', 'shop', 'brand', 'color', 'title', 'price', 'description',
    'image_urls'
]]
```

```python
em.set_key(zal, 'offer_id')
em.set_key(abt, 'offer_id')
```

```python
print('Number of tuples in Zalando: ' + str(len(zal)))
print('Number of tuples in Aboutyou: ' + str(len(abt)))
print('Number of tuples in A X B (i.e the cartesian product): ' +
      str(len(zal) * len(abt)))
```

```python
ob = em.OverlapBlocker()
K1 = ob.block_tables(zal,
                     abt,
                     'title',
                     'title',
                     l_output_attrs=[
                         'offer_id', 'shop', 'brand', 'color', 'price',
                         'title', 'description', 'image_urls'
                     ],
                     r_output_attrs=[
                         'offer_id', 'shop', 'brand', 'color', 'price',
                         'title', 'description', 'image_urls'
                     ],
                     overlap_size=2)
```

```python
K2 = ob.block_candset(K1, 'brand', 'brand', overlap_size=1)
```

```python
K3 = ob.block_candset(K2, 'color', 'color', overlap_size=1)
```

```python
print(len(K1))
print(len(K2))
print(len(K3))
```

```python
K3 = K3.sample(frac=1).reset_index(drop=True)
```

```python
def match_similar_priced(ltuple, rtuple):
    # assume that there is a 'name' attribute in the input tables
    # and each value in it has two words
    if pd.notna(ltuple['price']) and pd.notna(rtuple['price']):
        if pd.notna(ltuple['price']) < 100:
            l_price = int(ltuple['price'])
            r_price = int(rtuple['price'])
            if abs(l_price - r_price) >=20:
                return True
            else:
                return False
        else:
            l_price = int(ltuple['price'])
            r_price = int(rtuple['price'])
            if abs(l_price - r_price) >=40:
                return True
            else:
                return False
    else:
        return False

bb = em.BlackBoxBlocker()
bb.set_black_box_function(match_similar_priced)
T4 = bb.block_candset(T3)
```
